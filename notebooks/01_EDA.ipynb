```python
import zipfile
import os

# Find the zip file
zip_file = './dataset/archive.zip'

# Unzip it
with zipfile.ZipFile(zip_file, 'r') as z:
    z.extractall('.')
    print("Files extracted:")
    print(z.namelist())  # shows you what was inside

# Confirm files are there
print("\nFiles in current folder:")
print(os.listdir('.'))
```
