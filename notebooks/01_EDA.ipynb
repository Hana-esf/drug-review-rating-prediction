### Data Loading

```python
import os
import zipfile
import pandas as pd

# Defining the relative path to the local zip file
zip_path = os.path.join('..', 'dataset', 'archive.zip')

print(f"Looking for local dataset archive at: {zip_path}")

if os.path.exists(zip_path):
    with zipfile.ZipFile(zip_path, 'r') as zip_file:
        # Listing files inside the zip to make sure we know the exact names
        print("Files found inside zip:", zip_file.namelist())
        
        # Loading the training and test CSV files directly into Pandas
        df_train = pd.read_csv(zip_file.open('drugsComTrain_raw.csv'))
        df_test = pd.read_csv(zip_file.open('drugsComTest_raw.csv'))
    
    print(f"\nData loaded successfully!")
    print(f"Train set shape: {df_train.shape}")
    print(f"Test set shape: {df_test.shape}")
else:
    print("Error: Could not find 'archive.zip' in your 'dataset' folder. Please check your file placement.")

# Displaying the first 5 rows
df_train.head()
```

### Checking for Missing Values and Data Types

```python
print("--- Training Set Metadata & Missing Values ---")
print(df_train.info())
print("\nMissing values per column:")
print(df_train.isnull().sum())

print("\n--- Test Set Metadata & Missing Values ---")
print(df_test.isnull().sum())
```
