### Data Loading

```python
import os
import zipfile
import pandas as pd

# Defining the relative path to the local zip file
zip_path = os.path.join('..', 'dataset', 'archive.zip')

print(f"Looking for local dataset archive at: {zip_path}")

if os.path.exists(zip_path):
    with zipfile.ZipFile(zip_path, 'r') as zip_file:
        # Listing files inside the zip to make sure we know the exact names
        print("Files found inside zip:", zip_file.namelist())
        
        # Loading the training and test CSV files directly into Pandas
        df_train = pd.read_csv(zip_file.open('drugsComTrain_raw.csv'))
        df_test = pd.read_csv(zip_file.open('drugsComTest_raw.csv'))
    
    print(f"\nData loaded successfully!")
    print(f"Train set shape: {df_train.shape}")
    print(f"Test set shape: {df_test.shape}")
else:
    print("Error: Could not find 'archive.zip' in your 'dataset' folder. Please check your file placement.")

# Displaying the first 5 rows
df_train.head()
```

### Checking for Missing Values and Data Types

```python
print("--- Training Set Metadata & Missing Values ---")
print(df_train.info())
print("\nMissing values per column:")
print(df_train.isnull().sum())

print("\n--- Test Set Metadata & Missing Values ---")
print(df_test.isnull().sum())
```

### Analyzing the Rating Distribution

Our model's job is to predict the rating (0 to 10), so we need to see if the dataset is heavily skwed toward hight or low rating. This helps us to decide what loss functions and evaluation metrics to use later. 

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate absolute counts and percentages for each star rating
rating_counts = df_train['rating'].value_counts().sort_index()
rating_pct = df_train['rating'].value_counts(normalize=True).sort_index() * 100

distribution_df = pd.DataFrame({
    'Count': rating_counts,
    'Percentage (%)': rating_pct.round(2)
})
print("Target Rating Distribution:")
print(distribution_df)

# Visualize the distribution
plt.figure(figsize=(8, 5))
sns.barplot(x=distribution_df.index, y=distribution_df['Count'], palette='viridis')
plt.title('Distribution of Patient Ratings (1-10 Stars)')
plt.xlabel('Rating (Stars)')
plt.ylabel('Number of Reviews')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# I use .str.split().apply(len) to count spaces between words
df_train['word_count'] = df_train['review'].str.split().apply(len)

# Calculating key statistics (Mean, Median, Min, Max, and Percentiles)
print("--- Review Word Count Statistics ---")
print(df_train['word_count'].describe())

# Calculating specific percentiles to help us choose a deep learning cutoff
print("\n--- Length Percentiles ---")
print(f"80th percentile: {df_train['word_count'].quantile(0.80):.0f} words")
print(f"90th percentile: {df_train['word_count'].quantile(0.90):.0f} words")
print(f"95th percentile: {df_train['word_count'].quantile(0.95):.0f} words")

# Visualize the text length distribution
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(9, 5))
sns.histplot(df_train['word_count'], bins=50, kde=True, color='teal')
plt.title('Distribution of Review Word Counts')
plt.xlabel('Number of Words per Review')
plt.ylabel('Number of Reviews')
plt.axvline(df_train['word_count'].quantile(0.90), color='red', linestyle='--', label='90th Percentile')
plt.legend()
plt.grid(axis='both', alpha=0.3)
plt.show()
```
